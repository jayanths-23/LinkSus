```python
import pandas as pd

df = pd.read_csv('Dataset/phishing_site_urls.csv')
df.head()
```

```python
import re

df['clean_url'] = df['URL'].apply(lambda x: re.sub(r'^https?://(www\.)?', '', x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv = TfidfVectorizer(
    analyzer='char',
    ngram_range=(3,5),
    max_features=100000   # limit size (important)
)

text_features = cv.fit_transform(df['clean_url'])
```

```python
def extract_features(url):
    return [
        len(url),
        url.count('.'),
        url.count('-'),
        sum(c.isdigit() for c in url),

        int('login' in url),
        int('secure' in url),
        int('verify' in url),
        int('account' in url),
        int('update' in url),

        int(url.count('.') > 3),
        int('@' in url),
        int('//' in url),
        int(len(url.split('.')) > 3),
    ]
```

```python
import numpy as np

url_features = np.array([extract_features(u) for u in df['clean_url']])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
url_features_scaled = scaler.fit_transform(url_features)
```

```python
from scipy.sparse import hstack

X = hstack([text_features, url_features_scaled])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    X, df['Label'], test_size=0.2
)
```

```python
from sklearn.svm import LinearSVC

model = LinearSVC(max_iter=10000)
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
test_url = "bank.com-security-update.com"

text_vec = cv.transform([test_url])
url_feat = np.array([extract_features(test_url)])
url_feat = scaler.transform(url_feat)

combined = hstack([text_vec, url_feat])

print(model.predict(combined))
```
